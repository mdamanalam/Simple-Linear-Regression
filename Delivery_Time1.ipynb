#Model Bulding

```python
import statsmodels.formula.api as smf #import Libray
```

```python
model = smf.ols("Delivery_Time~Sorting_Time",data = data).fit()
```

```python
sns.regplot(x="Sorting_Time",y="Delivery_Time",data=data)  #Scatter Plot
```

```python
#cofficient
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)  
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

#The R^2=0.68 value are so less 

#We detect the outlier and remove it 

#Delete 20th row

```python
from pandas.io.formats.info import DataFrameTableBuilderNonVerbose
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
df=data.drop(data.index[[20]],axis=0).reset_index(drop=True)
df
```

```python
#Correlation
df.corr()
```

```python
model = smf.ols("Delivery_Time~Sorting_Time",data = df).fit()
```

```python
sns.regplot(x="Sorting_Time",y="Delivery_Time",data=df)  #Scatter Plot
```

```python
#cofficient
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)  
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

#The R^2=0.76
