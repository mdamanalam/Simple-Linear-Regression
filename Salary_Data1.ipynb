#Model Bulding

```python
import statsmodels.formula.api as smf #import Libray
```

```python
model = smf.ols("Salary~YearsExperience",data = data).fit()
```

```python
sns.regplot(x="YearsExperience",y="Salary",data=data)  #Scatter Plot
```

```python
#cofficient
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)  
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

#Model Predictions

```python
# Manual prediction for say sorting time 3
Salary = (25792.200199) + (9449.962321)*(3)
Salary
```

```python
# Automatic Prediction for say sorting time 3,5
Salary_Data=pd.Series([3,5])
Salary_Data
```

```python
data_pred=pd.DataFrame(Salary_Data,columns=['YearsExperience'])
data_pred
```

```python
model.predict(data_pred)
```
